```python
from struct import *
import asyncio
import nest_asyncio
nest_asyncio.apply()

async def binary_decode(reader):
    print("begin of binary_decode().")
    Type = await reader.readexactly(1)
    Type = int.from_bytes(Type, byteorder='big', signed=False)
    print("first byte read in binary_decode()")
    print(Type)
    
    Type_dict = {"BIN_LONGLIST": 0,
                    "BIN_INTEGER": 1,
                    "BIN_REAL": 2,
                    "BIN_BOOLEAN": 3,
                    "BIN_LONGSTRING": 4,
                    "BIN_LONGSYMBOL": 5,
                    "BIN_LONGTEXT": 6,
                    "BIN_LIST": 10,
                    "BIN_SHORTLIST": 11,
                    "BIN_SHORTINT": 12,
                    "BIN_BYTE":13,
                    "BIN_STRING": 14,
                    "BIN_SHORTSTRING": 15,
                    "BIN_SYMBOL": 16,
                    "BIN_SHORTSYMBOL": 17,
                    "BIN_TEXT": 18,
                    "BIN_SHORTTEXT": 19,
                    "BIN_DOUBLE": 20}
    typeCode = None
    for typ, val in Type_dict.items():
        if val == Type:
            typeCode = val
            break
    
    if typeCode == 0: #LongList
        print("LongList")
        LE = []
        barray = bytearray(4) ####readInt()
        barray = await reader.readexactly(4)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Integer-Byte read from Stream!"
        r = unpack('>I',barray)[0]
        ####
        if r == 0:
            return LE
        else:
            for i in range(r):
                LE.append(await binary_decode())
            return LE
            
    elif typeCode == 1:
        ####readInt()
        
        print("Int")
        barray = bytearray(4)
        barray = await reader.readexactly(4)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Integer-Byte read from Stream!"
        r = unpack('>I',barray)[0]
        ####
        print("int: ")
        print(r)
        return r 
        
    elif typeCode == 2:
        ####readReal()
        print("Real")
        barray = bytearray(4)
        barray = await reader.readexactly(4)
        print(barray)
        assert (int.from_bytes(barray, order = 'big', signed = True)) >= 0, "Invalid Real-Byte read from Stream!"
        r = unpack('>f',barray)[0]
        ####
        print("real: ")
        print(r)
        return r 
        
    elif typeCode == 3:
        ####readBool()
        print("Bool")
        barray = bytearray(1)
        barray = await reader.readexactly(1)
        print(barray)
        assert (int.from_bytes(barray, order = 'big', signed = True)) >= 0, "Invalid Boolean-Byte read from Stream!"
        r = unpack('?',barray)[0]
        ####
        print("bool: ")
        print(r)
        return r
        
            
    elif typeCode == 4:
        ####readLongString()
        print("LongString")
        barray = bytearray(4) #readInt()
        barray = await reader.readexactly(4)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid Int-Byte read from Stream!"
        len = unpack('>I',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid LongSymbol-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("long string: ")
        print(barray.decode())
        return '"' + barray.decode() + '"'
        
    elif typeCode == 5:
        ####readLongSymbol()
        print("LongSymbol")
        barray = bytearray(4) #readInt()
        barray = await reader.readexactly(4)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid Int-Byte read from Stream!"
        len = unpack('>I',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid LongSymbol-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("long symbol: ")
        print(barray.decode())
        return barray.decode()
        
    elif typeCode == 6:
        ####readLongText()
        print("LongText")
        barray = bytearray(4) #readInt()
        barray = await reader.readexactly(4)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid Int-Byte read from Stream!"
        len = unpack('>I',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid LongText-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("Long Text: ")
        txt = barray.decode()
        if not txt.startswith('<text>'):
            txt = '"<text>' + txt + '</text--->"'
            
        print(txt)
        return txt
        
    elif typeCode == 10: #List
        print("List")
        LE = []
        ####readShort()
        barray = bytearray(2)
        barray = await reader.readexactly(2)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, order = 'big', signed = True)) >= 0, "Invalid ShortInt-Byte read from Stream!"
        r = unpack('>H',barray)[0]
        ####
        if r == 0:
            return LE
        else:
            for i in range(r):
                LE.append(await binary_decode())
            return LE
        
    elif typeCode == 11: #ShortList
        print("ShortList")
        LE = []
        ####readByte()
        barray = bytearray(1)
        barray = await reader.readexactly(1)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Byte read from Stream!"
        r = int.from_bytes(barray, byteorder = 'big', signed = False)
        ####
        if r == 0:
            return LE
        else:
            for i in range(r):
                LE.append(await binary_decode())
            return LE
        
    elif typeCode == 12:
        ####readShort()
        print("readShort")
        barray = bytearray(2)
        barray = await reader.readexactly(2)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortInt-Byte read from Stream!"
        r = unpack('>H',barray)[0]
        print("short: ")
        print(r)
        return r
        
    elif typeCode == 13:
        ####readByte()
        print("Byte")
        barray = bytearray(1)
        barray = await reader.readexactly(1)
        print(barray)
        #assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Byte read from Stream!"
        r = int.from_bytes(barray, byteorder = 'big', signed = True)
        print("byte: ")
        print(r)
        return r
        
    elif typeCode == 14:
        ####readString()
        print("String")
        barray = bytearray(2)
        barray = await reader.readexactly(2)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortInt-Byte read from Stream!"
        len = unpack('>H',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid String-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("string: ")
        print(barray.decode())
        return '"' + barray.decode() + '"'
        
    elif typeCode == 15:
        ####readShortString()
        print("ShortString")
        barray = bytearray(1)
        barray = await reader.readexactly(1)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Byte read from Stream!"
        len = int.from_bytes(barray, byteorder = 'big', signed = False)
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortString-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("short string: ")
        print(barray.decode())
        return '"' + barray.decode() + '"'
    
    elif typeCode == 16:
        ####readSymbol()
        print("Symbol")
        barray = bytearray(2) #readShort()
        barray = await reader.readexactly(2)
        print(barray)
        for b in barray:
            assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortInt-Byte read from Stream!"
        len = unpack('>H',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid String-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("symbol: ")
        print(barray.decode())
        return barray.decode()
        
    elif typeCode == 17:
        ####readShortSymbol()
        print("ShortSymbol")
        barray = bytearray(1) #readByte()
        barray = await reader.readexactly(1)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Byte read from Stream!"
        len = int.from_bytes(barray, byteorder = 'big', signed = False)
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortString-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("short symbol: ")
        print(barray.decode())
        return barray.decode()
        
    elif typeCode == 18:
        ####readText()
        print("Text")
        barray = bytearray(2) #readShort()
        barray = await reader.readexactly(2)
        print(barray)
        #for b in barray:
         #   assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortInt-Byte read from Stream!"
        len = unpack('>H',barray)[0]
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid Text-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        txt = barray.decode()
        if not txt.startswith('<text>'):
            txt = '"<text>' + txt + '</text--->"'
            
        print(txt)
        return txt
        
        
    elif typeCode == 19:
        ####readShortText()
        print("ShortText")
        barray = bytearray(1) #readByte()
        barray = await reader.readexactly(1)
        print(barray)
        #assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Byte read from Stream!"
        len = int.from_bytes(barray, byteorder = 'big', signed = False)
        barray = bytearray(len)
        barray = await reader.readexactly(len)
        print(barray)
        #for b in barray:
            #assert (int.from_bytes(b, byteorder = 'big', signed = True)) >= 0, "Invalid ShortText-Byte read from Stream!"
        ###checking Environment.ENCODING not considered yet
        print("short Text: ")
        txt = barray.decode()
        if not txt.startswith('<text>'):
            txt = '"<text>' + txt + '</text--->"'
            
        print(txt)
        return txt
        
    elif typeCode == 20:
        ####readDouble()
        print("Double")
        barray = bytearray(8)
        barray = await reader.readexactly(8)
        print(barray)
        assert (int.from_bytes(barray, byteorder = 'big', signed = True)) >= 0, "Invalid Double-Byte read from Stream!"
        r = unpack('>d',barray)[0]
        ####
        print("Double: ")
        print(r)
        return r
```
