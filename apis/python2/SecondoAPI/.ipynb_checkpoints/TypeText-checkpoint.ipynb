```python
import import_ipynb
#import GeneralType
#from Abst_Repo import Parsable, Matchable

class TypeText():
    
    PREFIX = "<text>"
    SUFFIX = "</text--->"
    
    #def __init__(self, value, defined):
     #   super(TypeText, self).__init__(value, defined)
    
    
    def __init__(self, value = None, defined = None):
        if value is not None:
            self.value = value
        if defined is None or defined == True:
            self.defined = True
        else:
            self.defined = False
    
    def InFromList(self, NList):
        value = NList[0] if not NList[0].startswith(PREFIX) else NList[0][len(PREFIX):]
        value = value if not value.endswith(SUFFIX) else value[:-len(SUFFIX)]
        self.value = value
        
    def OutToList(self, NList):
        value = PREFIX + self.value + SUFFIX
        NList = []
        NList.append(value)
        return NList
    
    def __eq__(self, other):
        
        if other is None:
            return False
        assert isinstance(other, TypeText), "Type Error!"
        valueOther = other.getValue()
        if (self.value == valueOther):
            return True
        return False
    
    def __hash__(self):
        return hash(self.value)
    
    def __gt__(self, other):
        
        if other is None:
            return False
        assert isinstance(other, TypeText), "Type Error!"
        return other.value < self.value

    def __ge__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeText), "Type Error!"
        return not self.value < other.value
    
    def __lt__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeText), "Type Error!"
        return other.value > self.value
    
    def __le__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeText), "Type Error!"
        return not other.value < self.value
    
    
    def __str__(self):
        if self.value == None:
            return "(text undefined)"
        return "(text " + str(self.value) + ")"
    
    def parse(text):
        if text == None:
            return None
        result = TypeText(text)
        return result
```
