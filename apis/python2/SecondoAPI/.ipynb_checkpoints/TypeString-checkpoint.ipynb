```python
import import_ipynb
#import GeneralType
#from Abst_Repo import Parsable, Matchable

class TypeString():
    
    #def __init__(self, value, defined):
     #   super(TypeString, self).__init__(value, defined)
    
    def __init__(self, value = None, defined = None):
        if value is not None:
            self.value = value
        if defined is None or defined == True:
            self.defined = True
        else:
            self.defined = False
    
    def InFromList(self, NList):
        self.value = NList[0]
        
    def OutToList(self, NList):
        NList = []
        NList.append(self.value)
        return NList
        
    
    
    def __eq__(self, other):
        
        if other is None:
            return False
        assert isinstance(other, TypeString), "Type Error!"
        valueOther = other.getValue()
        if (self.value == valueOther):
            return True
        return False
    
    def __hash__(self):
        return hash(self.value)
    
    def __gt__(self, other):
        
        if other is None:
            return False
        assert isinstance(other, TypeString), "Type Error!"
        return other.value < self.value

    def __ge__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeString), "Type Error!"
        return not self.value < other.value
    
    def __lt__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeString), "Type Error!"
        return other.value > self.value
    
    def __le__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeString), "Type Error!"
        return not other.value < self.value
    
    
    def __str__(self):
        if self.value == None:
            return "(string undefined)"
        return "(String " + str(self.value) + ")"
    
    def parse(text):
        if text == None:
            return None
        result = TypeString(text)
        return result
    
```
