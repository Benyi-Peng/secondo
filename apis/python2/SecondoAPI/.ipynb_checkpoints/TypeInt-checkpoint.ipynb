```python
import import_ipynb
#from Abst_Repo import Parsable, Matchable, Summable
#import GeneralType

class TypeInt():
    
    #def __init__(self, value, defined):
     #   super(TypeInt, self).__init__(value, defined)
    
    def __init__(self, value = None, defined = None):
        if value:
            self.value = value
        if not defined or defined == True:
            self.defined = True
        else:
            self.defined = False
    
    def InFromList(self, NList):
        self.value = NList[0]
        
    def OutToList(self, NList):
        NList = []
        NList.append(self.value)
        return NList
    
    
    def __eq__(self, other):
        
        if not other:
            return False
        assert isinstance(other, TypeInt), "Type Error!"
        valueOther = other.getValue()
        if (self.value == valueOther):
            return True
        return False
    
    def __hash__(self):
        return hash(self.value)
    
    def __gt__(self, other):
        
        if other is None:
            return False
        assert isinstance(other, TypeInt), "Type Error!"
        return other.value < self.value

    def __ge__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeInt), "Type Error!"
        return not self.value < other.value
    
    def __lt__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeInt), "Type Error!"
        return other.value > self.value
    
    def __le__(self, other):
        if other is None:
            return False
        assert isinstance(other, TypeInt), "Type Error!"
        return not other.value < self.value
    
    def __add__(self,other):
        assert isinstance(other, TypeInt), "Type Error!"
        result = TypeInt(self.value + other.value)
        return result
            
    
    def __str__(self):
        return "(int " + str(self.value) + ")"
    
    def parse(text):
        value = int(text)
        result = TypeInt(value)
        return result
    
    def toFloat(self):
        return float(self.value)
   
    def setValue(self,value):
        self.value = value
    
    def getValue(self):
        return self.value
    
    def setDefined(self,value):
        self.defined = defined
    
    def getDefined(self):
        return self.defined
    
```
