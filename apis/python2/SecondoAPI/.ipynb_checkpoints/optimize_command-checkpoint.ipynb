```python
async def opt_comm_exec(reader, writer, query, db_name, flag):
    
    if reader is None or writer is None:
        print("Connection Error!")
        return None
    else:
        
        #execute query
        if flag:
            writer.write(b'<execute>\n')
        
        #optimize query
        else: 
            writer.write(b'<optimize>\n')
    
        writer.write(b'<database>\n')
        db_name += "\n"
        writer.write(db_name.encode())
    
        writer.write(b'</database>\n')
    
        writer.write(b'<query>\n')
        query += "\n"
        writer.write(query.encode())
    
        writer.write(b'</query>\n')
    
        if flag:
            writer.write(b'</execute>\n')
        
        else: 
            writer.write(b'</optimize>\n')
    
        await writer.drain()
    
        res = await reader.readline()
        res = res.decode()
        print("first response from OptServer: ")
        print(res)
        if not res:
            print("Connection to Optimizer is broken!")
            return None
        if res != "<answer>\n":
            print("Protocol Error!")
            return None
        #res = await optrdr.readuntil(b'</answer>')
        res = await reader.readline()
        print("second response from OptServer: ")
        print(res)
        res = res.decode()
        if not res:
            print("Connection to Optimizer is broken!")
            return None
        opt_res = ""
        count = 3
        while res != "</answer>\n":
            if flag:
                opt_res += res
            else:
                opt_res += res.replace(" \n", "")
            res = await reader.readline()
            print(str(count) + "th response from OptServer: ")
            print(res)
            res = res.decode()
            count += 1
            if not res:
                print("Connection to Optimizer is broken!")
                return None
    
        if opt_res == "":
            print("Optimization failed!")
            return None
        print(opt_res)
        return opt_res
```
