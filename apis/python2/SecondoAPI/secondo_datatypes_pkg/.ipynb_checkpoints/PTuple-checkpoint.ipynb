```python
import import_ipynb
#import PGenType
import Rel_Attrib

class PTuple():
    
    def __init__(self, attrib_list = None, value_list = None):
        #list of objects of types PInt, PReal, ...
        self.attrib_list = []
        if attrib_list is not None:
            self.attrib_list = attrib_list
        #list of types of objects
        self.value_list = []
        if value_list is not None:
            self.value_list = value_list
    
    def add_value(self, val):
        self.value_list.append(val)
        
    def get_value(self, index):
        return self.value_list[index]
    
    def get_value_list(self):
        return self.value_list
    
    def set_value_list(self, ls):
        self.value_list = ls
    
    def add_attrib(self, attrib):
        self.attrib_list.append(attrib)
        
    def get_attrib(self, index):
        return self.attrib_list[index]
    
    def get_attrib_list(self):
        return self.attrib_list
    
    def set_attrib_list(self, ls):
        self.attrib_list = ls
    
    def __eq__(self, obj):
        if obj == self:
            return True
        if obj is None:
            return False
        if not isinstance(obj, PTuple):
            return False
        otherTuple = PTuple()
        otherTuple = obj
        if not(len(self.value_list) == len(otherTuple.get_value_list())):
            return False
        for idx, att in enumerate(self.value_list):
            if not(att == otherTuple.get_value(idx)):
                return False
        for idx, att in enumerate(self.attrib_list):
            if not(att == otherTuple.get_attrib(idx)):
                return False
        
        return True

    # usage: tuple = PTuple.create_tuple_instance(attrib_list)
    @classmethod
    def create_tuple_instance(cls, attrib_list):
         return cls(attrib_list)
```
